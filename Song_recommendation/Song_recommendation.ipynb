```python
def rating(rating):
  rating = str(rating)
  rating = rating.split('/')
  rating = rating[0]
  # return float(rating)
  return rating

rating('9.9/10')
```

```python
df['User-Rating'] = df['User-Rating'].apply(rating)
```

```python
df['Album/Movie'] = df['Album/Movie'].str.replace(' ','')
df['Singer/Artists'] = df['Singer/Artists'].str.replace(',',' ').replace(' ','')
```

```python
df.head()
```

```python
df['tags'] = df['Singer/Artists'] + ' ' + df['Genre'] + ' ' + df['Album/Movie'] + ' ' + df['User-Rating']
df['tags'] = df['tags'].apply(lambda x: x.lower())
df['tags'][0]
```

```python
new_df = df[['Song-Name', 'tags']]
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=2000)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
# cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])
```

```python
def recommend(music):
  music_index = new_df[new_df['Song-Name'] == music].index[0]
  distances = similarity[music_index]
  music_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

  for i in music_list:
    print(new_df.iloc[i[0]]['Song-Name'])
```

```python
recommend('Tere Bin')
```

```python
recommend('Hawa Hawa')
```

```python
recommend('Mungda')
```
